# What I learned

### 1. SNR (signal to noise ratio)

$$\text{SNR} \triangleq \frac{\mathbb{E}[Z^2]}{\mathbb{E}[\epsilon^2]} = \frac{\Sigma_0 + \mu_0^2 }{\Sigma_y}$$


recurrence formula:
$$SNR_{(n+1)} = \frac{SNR_{(n)}}{1 + SNR_{(n)}}$$ $$$$
before-after relation:
$$\frac{1}{SNR_{(n+1)}} = \frac{1}{SNR_{(n)}} + 1$$
continued fraction form:
$$SNR_{(n)} = \frac{1}{1 + \frac{1}{1 + \frac{1}{\ddots + \frac{1}{SNR_{(1)}}}}}$$


### 2. posterior mean update

$$\mu_n = \mu_{n-1} + \frac{SNR_{(n)}}{1 + SNR_{(n)}} (y_n - \mu_{n-1})$$

#### kalman gain (scalar):
$$\frac{SNR_{(n)}}{1 + SNR_{(n)}}$$
