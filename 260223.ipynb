# What I learned

##- Scalar estimation

### 1. posterior precision

posterior precision: $\lambda_N = \lambda_0 + N\lambda_y$

weighted sum of prior and data means: $\mu_n = \frac{N\lambda_y \bar{y} + \lambda_0 \mu_0}{\lambda_N} = \frac{N\lambda_y}{N\lambda_y + \lambda_0} \bar{y} + \frac{\lambda_0}{N\lambda_y + \lambda_0} \mu_0$


###2. setting the prior mean ($\mu_0$)


        prior mean is Hyperparameter.
        there are 3 ways to determine this.

1. uninformative Prior


        set this parameter to 0 and bring the initial accuracy closer to 0.
        meaning: system believes only the data.

2. baseline


        use the average of reference data before data collection.
        In BCI, the user's Resting State signal is measured.

3. recursive


        the previous posterior mean is used as the prior mean.
        knowledge accumulates as data is updated.

###3. posterior mean

$$\mu_1 = \frac{\Sigma_y}{\Sigma_y + \Sigma_0} \mu_0 + \frac{\Sigma_0}{\Sigma_y + \Sigma_0} y \text{}$$
$$= \mu_0 + (y - \mu_0) \frac{\Sigma_0}{\Sigma_y + \Sigma_0} \text{}$$
$$= y - (y - \mu_0) \frac{\Sigma_y}{\Sigma_y + \Sigma_0} \text{}$$
