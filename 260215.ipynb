확률분포 그리기

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.normal(0, 1, 1000) #generate 100 sample data; avg: 0, std: 1

plt.hist(data, bins=30, density = True, alpha=1, color='g')
plt.title("Gaussian Distribution")
plt.show()
```

정규분포와 랜덤생성 분포 비교

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mu, sigma = 0, 1
data = np.random.normal(mu, sigma, 1000) # 1000 sample data, avg:0, std:2

#drawing histogram
count, bins, ignored = plt.hist(data, bins=30, density=True, alpha=0.6, color='g', label='Actual Data')

# normal distribution pdf
plt.plot(bins, norm.pdf(bins, mu, sigma), color='r', lw=2, label='Theoretical PDF')

plt.title(f"EEG Signal Distribution (mu={mu}, sigma={sigma})")
plt.legend() # show legend
plt.show()
```
