확률분포 그리기

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.normal(0, 1, 1000) #generate 100 sample data; avg: 0, std: 1

plt.hist(data, bins=30, density = True, alpha=1, color='g')
plt.title("Gaussian Distribution")
plt.show()
```

정규분포와 랜덤생성 분포 비교

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mu, sigma = 0, 1
data = np.random.normal(mu, sigma, 1000) # 1000 sample data, avg:0, std:2

#drawing histogram
count, bins, ignored = plt.hist(data, bins=30, density=True, alpha=0.6, color='g', label='Actual Data')

# normal distribution pdf
plt.plot(bins, norm.pdf(bins, mu, sigma), color='r', lw=2, label='Theoretical PDF')

plt.title(f"EEG Signal Distribution (mu={mu}, sigma={sigma})")
plt.legend() # show legend
plt.show()
```

Comparison of Distributions

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, laplace, t

# 1. Set the x-axis range from -10 to 10
x = np.linspace(-10, 10, 1000)

# 2. Calculate Probability Density Functions (PDF)
# Gaussian (Mean=0, Scale=1.5)
gaussian_pdf = norm.pdf(x, loc=0, scale=1.5)

# Laplace (Location=0, Scale=1.5) - Related to L1 Regularization (Sparsity)
laplace_pdf = laplace.pdf(x, loc=0, scale=1.5)

# Student-t (df=1, Location=0, Scale=1.5) - Heavy-tailed for Robustness
# As df increases, it converges to a Gaussian distribution
t_pdf = t.pdf(x, df=1, loc=0, scale=1.5)

# 3. Visualization
plt.figure(figsize=(10, 6))

plt.plot(x, gaussian_pdf, label='Gaussian (Normal)', color='blue', lw=2)
plt.plot(x, laplace_pdf, label='Laplace (L1 Prior)', color='red', lw=2)
plt.plot(x, t_pdf, label='Student t (Robust)', color='green', lw=2, linestyle='--')

plt.title("Comparison of Probability Distributions")
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
